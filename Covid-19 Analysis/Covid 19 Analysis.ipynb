Case Study: Covid 19 Data Analysis

Purpose: to find the relation between covid cases and happiness report of countries

```python
#import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
dataset.columns
```

```python
#delete the useless columns
df = dataset.drop(["Lat","Long"], axis = 1, inplace = True)
dataset.head()
```

```python
#aggregate the rows by the country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
corona_dataset_aggregated
```

```python
corona_dataset_aggregated.shape
```

```python
#visualise data related to a country
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
#calculate a good measure
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
#calculate the first derivative of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#maximum infection rate
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
#maximum infection rates for all the countries
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rates"]=max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
#create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
corona_data
```

```python
#importing the dataset 2
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report
```

```python
#drop the useless columns
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report.drop(useless_cols,axis=1,inplace=True)
happiness_report
```

```python
happiness_report.set_index("Country or region", inplace=True)
happiness_report.head()
```

```python
#join the dataset
corona_data.shape
```

```python
happiness_report.shape
```

```python
data = corona_data.join(happiness_report, how = "inner")
data
```

```python
data.corr()
```

```python
#Visualization
x = data["GDP per capita"]
y = data["Max_infection_rates"]
sns.scatterplot(x=x,y=np.log(y))#logscaling
plt.xlabel("GDP per capita")
plt.ylabel("Max Infection Rates")
plt.title("Scatter Plot: GDP per capita vs Max Infection Rates")
plt.show()
```

```python
sns.regplot(x=x,y=np.log(y))
plt.xlabel("GDP per capita")
plt.ylabel("Max Infection Rates")
plt.title("Regression Plot: GDP per capita vs Max Infection Rates")
plt.show()
```
